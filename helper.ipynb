```python
import pandas as pd
from IPython.display import display, HTML
import numpy as np
file = 'outputs/analysis/multiclass_both.csv'
pd.set_option('display.precision', 2)
def get_top_features(df, cols = ['feature', 'mean_percent', 'p']):

    for col in cols:
        
        display(HTML(df[[col]].to_html(index=False)))

```

```python
files = ['outputs/analysis/HC_vs_ADMCI_both.csv', 
         'outputs/analysis/HC_vs_FTD_both.csv', 
         'outputs/analysis/HC_vs_PD_both.csv',
         'outputs/analysis/HC_vs_ALS_both.csv',
         'outputs/analysis/fs_ADMCI_vs_FTD_both.csv',
         'outputs/analysis/fs_ADMCI_vs_PD_both.csv',
         'outputs/analysis/fs_FTD_vs_ALS_both.csv',]
multiclass = pd.read_csv('outputs/analysis/multiclass_both.csv')
multiclass['mean_percent'] = np.nan
multiclass= multiclass.head(3)


combined = pd.concat([pd.read_csv(f).head(3) for f in files])
combined = pd.concat([multiclass, combined, ])
combined = combined[['feature', 'mean_percent', 'p']]
```

```python
combined.to_csv('outputs/analysis/predictor_quick_table.csv', index=False)
```

```python
get_top_features(combined)
combined
```
